```
# -*- coding: utf-8 -*-
"""jogos_Hallowen.ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1eTmFvX3ypAXUVDlbDSIDUMo1thQS4bJ9
"""

import random
import time
import os

# -----------------------------
# Fun√ß√µes de utilidade
# -----------------------------
def limpar_tela():
    os.system('cls' if os.name == 'nt' else 'clear')

def print_lento(texto, atraso=0.03):
    for letra in texto:
        print(letra, end='', flush=True)
        time.sleep(atraso)
    print()

def linha():
    print("üï∏Ô∏è" * 40)

# -----------------------------
# Jogo 1 - Forca Assombrada
# -----------------------------
def forca_assombrada():
    limpar_tela()
    linha()
    print("üéÉ  FORCA ASSOMBRADA  üéÉ")
    linha()
    print("\nAdivinhe a palavra antes que o monstro te encontre...\n")

    palavras = {
        "fantasma": "Assombra castelos antigos...",
        "bruxa": "Adora voar em noites de lua cheia.",
        "abobora": "S√≠mbolo cl√°ssico do Halloween.",
        "zumbi": "Voltou dos mortos e anda bem devagar.",
        "aranha": "Costuma morar nos cantos escuros.",
        "caveira": "Representa a morte... e o mist√©rio."
    }

    palavra, dica = random.choice(list(palavras.items()))
    letras_descobertas = ["_" for _ in palavra]
    tentativas = 6
    usadas = set()

    print(f"Dica: {dica}\n")

    def mostrar_forca(erros):
        estagios = [
            "   -----\n   |   |\n   |   O\n   |  /|\\\n   |  / \\\n  _|_",
            "   -----\n   |   |\n   |   O\n   |  /|\\\n   |  /\n  _|_",
            "   -----\n   |   |\n   |   O\n   |  /|\\\n   |\n  _|_",
            "   -----\n   |   |\n   |   O\n   |  /|\n   |\n  _|_",
            "   -----\n   |   |\n   |   O\n   |   |\n   |\n  _|_",
            "   -----\n   |   |\n   |   O\n   |\n   |\n  _|_",
            "   -----\n   |   |\n   |\n   |\n   |\n  _|_"
        ]
        print(estagios[6 - erros])

    while tentativas > 0 and "_" in letras_descobertas:
        print("Palavra:", " ".join(letras_descobertas))
        print("Letras usadas:", " ".join(sorted(usadas)) or "(nenhuma)")
        mostrar_forca(tentativas)
        palpite = input("\nDigite uma letra: ").lower().strip()

        if not palpite.isalpha() or len(palpite) != 1:
            print("Digite apenas uma letra v√°lida!")
            continue
        if palpite in usadas:
            print("Voc√™ j√° tentou essa letra.")
            continue

        usadas.add(palpite)
        limpar_tela()
        if palpite in palavra:
            for i, c in enumerate(palavra):
                if c == palpite:
                    letras_descobertas[i] = palpite
            print("üéÉ Boa! Voc√™ acertou uma letra!")
        else:
            tentativas -= 1
            print("üëª Letra errada! O monstro est√° chegando...")
        print()
        time.sleep(1)

    if "_" not in letras_descobertas:
        print(f"\n‚ú® Voc√™ escapou! A palavra era **{palavra.upper()}**! ‚ú®")
    else:
        mostrar_forca(0)
        print(f"\nüíÄ Voc√™ perdeu! A palavra era **{palavra.upper()}**... O monstro te pegou!")

    input("\nPressione Enter para voltar ao menu...")

# -----------------------------
# Jogo 2 - Adivinha dos Doces
# -----------------------------
def adivinha_doces():
    limpar_tela()
    linha()
    print("üç¨  ADIVINHA DOS DOCES M√ÅGICOS  üç¨")
    linha()
    print("\nUma bruxa escondeu doces em um caldeir√£o encantado...\n")

    nivel = input("Escolha a dificuldade (f√°cil / m√©dio / dif√≠cil): ").lower()
    if nivel == "facil":
        limite, tentativas = 20, 6
    elif nivel == "medio":
        limite, tentativas = 40, 5
    else:
        limite, tentativas = 80, 4

    numero_secreto = random.randint(1, limite)
    print(f"\nA bruxa cochicha: 'H√° entre 1 e {limite} doces...'")

    while tentativas > 0:
        try:
            palpite = int(input(f"\nVoc√™ tem {tentativas} tentativas. Adivinhe: "))
        except ValueError:
            print("Digite um n√∫mero v√°lido!")
            continue

        if palpite == numero_secreto:
            print("\n‚ú® CORRETO! Voc√™ ganhou todos os doces! üç≠‚ú®")
            break
        elif palpite < numero_secreto:
            print("üîÆ A bruxa ri: 'Mais... bem mais doces que isso!'")
        else:
            print("üíÄ Ela sussurra: 'Menos... voc√™ exagerou!'")

        tentativas -= 1
        time.sleep(1)

    if tentativas == 0 and palpite != numero_secreto:
        print(f"\nA po√ß√£o evapora... o n√∫mero era {numero_secreto}! Nenhum doce pra voc√™! üï∏Ô∏è")

    input("\nPressione Enter para voltar ao menu...")

# -----------------------------
# Jogo 3 - A Casa do Fantasma
# -----------------------------
def aventura_fantasma():
    limpar_tela()
    linha()
    print("üèöÔ∏è  A CASA DO FANTASMA üïØÔ∏è")
    linha()
    print()
    print_lento("Voc√™ e seus amigos entraram em uma velha mans√£o para ca√ßar doces...")
    print_lento("Mas as portas se fecharam sozinhas!")
    print_lento("Agora, voc√™ precisa encontrar a sa√≠da antes que o fantasma apare√ßa...\n")

    escolha = input("H√° tr√™s portas: esquerda, centro e direita. Qual voc√™ escolhe? ").lower()
    if escolha == "esquerda":
        print_lento("\nVoc√™ entra em uma biblioteca antiga.")
        acao = input("Quer acender uma vela ou sair? (acender/sair): ").lower()
        if acao == "acender":
            print_lento("A vela acende e revela uma passagem secreta!")
            print_lento("Voc√™ escapa pela passagem e vence! üéâ")
        else:
            print_lento("Um livro cai... um esp√≠rito sai dele e te persegue! üíÄ Fim!")
    elif escolha == "centro":
        print_lento("\nVoc√™ sobe as escadas que rangem...")
        acao = input("Entrar no quarto iluminado ou no escuro? (iluminado/escuro): ").lower()
        if acao == "iluminado":
            print_lento("Voc√™ encontra doces e um bilhete: 'Voc√™ est√° salvo!' üç¨ Vit√≥ria!")
        else:
            print_lento("Um espelho se quebra sozinho... e algo sai dele! üëª Fim!")
    elif escolha == "direita":
        print_lento("\nVoc√™ entra na cozinha. Um caldeir√£o borbulha no fog√£o...")
        acao = input("Olhar dentro ou fugir? (olhar/fugir): ").lower()
        if acao == "olhar":
            print_lento("Um bra√ßo sai do caldeir√£o e te puxa! üíÄ Fim!")
        else:
            print_lento("Voc√™ foge pelos fundos e escapa! üéÉ Vit√≥ria!")
    else:
        print_lento("Voc√™ hesitou demais... o fantasma apareceu! üëª Fim!")

    input("\nPressione Enter para voltar ao menu...")

# -----------------------------
# Menu principal
# -----------------------------
def menu():
    while True:
        limpar_tela()
        linha()
        print("üéÆ  PACOTE DE JOGOS DE HALLOWEEN üéÆ")
        linha()
        print("\n1 - Forca Assombrada")
        print("2 - Adivinha dos Doces M√°gicos")
        print("3 - A Casa do Fantasma")
        print("0 - Sair")
        linha()
        opcao = input("\nEscolha um jogo: ")

        if opcao == "1":
            forca_assombrada()
        elif opcao == "2":
            adivinha_doces()
        elif opcao == "3":
            aventura_fantasma()
        elif opcao == "0":
            limpar_tela()
            print("üëª Obrigado por jogar! Feliz Halloween! üéÉ")
            break
        else:
            print("Op√ß√£o inv√°lida!")
            time.sleep(1)

# -----------------------------
# Execu√ß√£o principal
# -----------------------------
if __name__ == "__main__":
  menu ()
```
